```python
import numpy as np
```

```python
x1 =np.random.normal(30, 3, 100) # random variable (mean = 30 ,std = 3 , total points =100)
x2 = np.random.normal(20, 2, 100)
x3 = np.random.normal(10, 3, 100)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x1,label='plot')
plt.plot(x2,label='2nd plot')
plt.plot(x3,label= '3rd plot')
plt.legend(bbox_to_anchor=(0.,1.02,1.,.102),
           loc = 3,
           ncol = 3,
           mode = "expand",
           borderaxespad = 0.5
           
          )


plt.annotate("Point the stick here ",(55,20),
             xycoords='data',
             xytext=(5, 35),
             arrowprops=dict(arrowstyle='->')
            )
plt.show()
```

```python
mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 10000)
```

```python
ax = plt.gca()
ax.hist(x,bins=35,color='b')
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d$' % (mu,
sigma))
plt.show()
```

```python
x = np.arange(0, 10, 1)
y = np.log(x)
xe = 0.1 * np.abs(np.random.randn(len(y)))
```

```python
plt.bar(x, y, yerr=xe,
        width=0.4,
        align='center', 
        ecolor='r',
        color='cyan', 
        label='experiment #1')

plt.xlabel('# measurement')
plt.ylabel('Measured values')
plt.title('Measurements')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.random.randn(1000)
y1 = np.random.randn(len(x))
y2 = 1.2 + np.exp(x)
```

```python
plt.figure(figsize=(20,8))
ax1 = plt.subplot(121)
plt.scatter(x, y1, color='indigo', alpha=0.3, edgecolors='white',
label='no correl')
plt.xlabel('no correlation')
plt.grid(True)
ax2 = plt.subplot(122, sharey=ax1, sharex=ax1)
plt.scatter(x, y2, color='green', alpha=0.3, edgecolors='grey',
label='correl')
plt.xlabel('strong correlation')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(1, figsize=(6,6))
ax = plt.axes([0.1, 0.1, 0.8, 0.8]) # creating own axis 

labels = 'Spring', 'Summer', 'Autumn', 'Winter'

x = [15, 30, 45, 10]

explode=(0.1, 0.1, 0.1, 0.1)

# remove the explode to see what explode does 
plt.pie(x,     # my data x
        explode=explode, # to blow up the pie chart , for better visibility
        labels=labels,  # labes for each section 
        autopct='%1.1f%%',  # show percentage value floating points
        startangle=67    # starting angle / change and see the impact 
       )

plt.title('Rainy day by seasons')
plt.show()
```
